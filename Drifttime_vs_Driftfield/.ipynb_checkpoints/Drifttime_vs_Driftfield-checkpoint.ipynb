## Import and Data Processing

```python
ctx = straxbra.XebraContext()
```

```python
run = '00201'
peaks_data = ctx.get_array(run, 'peaks')
```

```python
waveform = peaks_data['data']
integral = peaks_data['area']
width = peaks_data['width'][:,5]
timestamp = peaks_data['time']
area_per_channel = peaks_data['area_per_channel']
risetime = -peaks_data['area_decile_from_midpoint'][:,1]
```

```python
def CutCondition(waveform):
    
    
    FirstPeak_area = sum(waveform[:15]) # range in samples (not ns!)
    SecondPeak_area = sum(waveform[15:]) # range in samples (not ns!)

    
    if  FirstPeak_area > 100 and FirstPeak_area < 600 and SecondPeak_area > 40 and SecondPeak_area  < 200: 
#        print (i , secondS1_area, secondS1_peaktime)
        return(True)
    else:
        return(False)
    
```

```python
def PrintWaveform(waveform,i):

    fig, (ax1, ax2) = plt.subplots(1, 2, sharex=False, figsize=(12, 3.5))

    ax1.plot(waveform[i])
    ax2.plot(waveform[i+1])
#     ax1.set_xlim([0, 50])
    plt.show()
```

```python
def S2_found(integral, width):
    if integral > 500 and width >200 and width < 2000:
        return True
    else:
        return False
```

```python
def Process(waveform, i):
    CutFlag = CutCondition(waveform[i])
    if CutFlag == True:
        Timestamps_kr_events.append(timestamp[i])
        j = 1
        counter_timestamp = timestamp[i+j]
        threshold_timestamp = timestamp[i] + 65000
        while threshold_timestamp >= counter_timestamp and S2_found(integral = integral[i+j], width = width[i+j]) == False: 
            j += 1
            counter_timestamp = timestamp[i+j]           
        else:
            if S2_found(integral = integral[i+j], width = width[i+j]):
                    #PrintWaveform(waveform[i+j])
                kr_s2_area.append(integral[i+j])
                kr_s2_width.append(width[i+j])
                kr_s2_risetime.append(risetime[i+j])
                kr_s2_drifttime.append(counter_timestamp - timestamp[i])
                    
                return
            else:
                return
```

```python
num = len(waveform)

kr_s2_area = []
kr_s2_width = []
kr_s2_risetime = []
kr_s2_drifttime = []
Timestamps_kr_events = []
Time_between_S1 = []
print_counter = 0.0
for i in range(0,num):
    
    Process(waveform,i)
    
    if i/num >= print_counter:
        print(round(print_counter,2), end = ' ')
        print_counter += 0.1
        
print(1.0)
```

## Driftvelocity vs Driftfield

```python
def power_law(x,a,b):
    power_law = a*x**b
    return power_law
```

```python
Drift_field = [500,450,400,350,300,250,200,150,100,50]
Drift_time_list = np.array([40500,41500,42500,43000,43800,45000,46500,48000,50000,57000])*1e-3
Drift_time_list_error = Drift_time_list*0.02

Xenon10_v = 1.880
Xenon100_v = 1.739
PandaXII_v = 1.722
Lux_v = 1.501
Xenon1T_v = [1.370,1.441,1.455,1.481]

Xenon10_field = 729
Xenon100_field = 528
PandaXII_field = 397
Lux_field = 180
Xenon1T_field = [92,124,134,154]

Drift_velocity_list = 71.5/np.array(Drift_time_list)
Drift_velocity_list_error = Drift_velocity_list*0.02

p0 = [0.7,0.1]
popt, pcov  = curve_fit(power_law, Drift_field, Drift_velocity_list, p0=p0, sigma=np.maximum(np.sqrt(Drift_velocity_list), np.ones_like(Drift_velocity_list)))

print(popt)
print(round(popt[0],2))
print(round(popt[1],2))
print(np.sqrt(np.diag(pcov)))
```

```python
#############################################
## Set plot styles:
#############################################

fig=plt.figure(figsize=(16, 10), dpi=150)

# Text options
#plt.rc('text', usetex=False)
plt.rc('font', family='serif')

# Labels
label_size = 32.5 #25
label_pad = 13 #10
title_size = 39 #30
title_pad = 26 #20

# Axes Parameters
axes = plt.gca()
axes.tick_params(axis='both') 

# Axes ---> Linewidth
for axis in ['top','bottom','left','right']:
    axes.spines[axis].set_linewidth(2.2)
    
# Axes ---> Ticks
axes.xaxis.set_minor_locator(AutoMinorLocator(5)) 
plt.tick_params(which='minor', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=6)
axes.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.tick_params(which='minor', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=6)
plt.tick_params(which='major', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=12)

# Grid
plt.grid(alpha = 0.5)
    
#############################################
## Plot specific settings and plot data
#############################################

# Data plot
Drift_field_range = np.linspace(20,800,100)

plt.errorbar(Drift_field,Drift_velocity_list,yerr=Drift_velocity_list_error,marker = 'x',fmt='x', c = 'k',markersize='11', ecolor='k',capsize=4, label = 'XeBRA')

plt.scatter(Xenon10_field, Xenon10_v, marker = 'o', s=200, c = 'C0', label = 'XENON10')
plt.scatter(Xenon100_field, Xenon100_v, marker = 'v', s=200, c = 'C1', label = 'XENON100')
plt.scatter(PandaXII_field, PandaXII_v, marker = 's', s=200, c = 'C2', label = 'PandaX-II')
plt.scatter(Lux_field, Lux_v, marker = '^', s=200, c = 'C3', label = 'LUX')
plt.scatter(Xenon1T_field, Xenon1T_v, marker = 'd', s=200, c = 'C4', label = 'XENON1T')

plt.plot(Drift_field_range,power_law(Drift_field_range,*popt), color = 'dimgrey', linewidth = 2.2, label = '$0.74 \cdot \mathrm{E}^{0.14}$')

plt.ylabel('Drift velocity [mm/$\mathrm{\mu}$s]', fontsize = label_size, labelpad = label_pad)
plt.xlabel('Drift field [V/cm]', fontsize = label_size, labelpad = label_pad)

#plt.yscale('log')
#plt.xscale('log')

plt.ylim(1.2, 2)
plt.xlim(0, 800)

# Legend
leg = plt.legend(loc=2, bbox_to_anchor=(1.01,1.025), fontsize=0.9*label_size, frameon=True, shadow=False, edgecolor='black', fancybox=False)
for line in leg.get_lines(): # enumerated
    line.set_linewidth(4)
leg.get_frame().set_linewidth(2.5)

fig.savefig("Drift_velocity_vs_drift_field.png", bbox_inches='tight')

plt.show()
```
