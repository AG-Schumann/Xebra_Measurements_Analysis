```python
def CutCondition(waveform, i):
    
    FirstPeak_area = sum(waveform[i][:15]) # range in samples (not ns!)
    Risetime_first_peak = np.argmax(waveform[i][:15])*10
    SecondPeak_area = sum(waveform[i][15:]) # range in samples (not ns!)
    Risetime_second_peak = np.argmax(waveform[i][15:])*10 +150
    
    if  FirstPeak_area > 100 and FirstPeak_area < 500 and SecondPeak_area > 40 and SecondPeak_area  < 150: 
        return(True, Risetime_first_peak, Risetime_second_peak)
    else:
        return(False, 0, 0)
    
```

```python
def Sum(area_per_channel, i):
    PMT_sum = 0.5*sum(area_per_channel[i][1:8])
    return PMT_sum
```

### Kr-83m lifetime

```python
def expo(x, a, b):
    return a*np.exp(-b*x)
```

```python
#############################################
## Set plot styles:
#############################################

fig=plt.figure(figsize=(16, 10), dpi=150)

# Text options
#plt.rc('text', usetex=False)
plt.rc('font', family='serif')

# Labels
label_size = 30 #25
label_pad = 12 #10
title_size = 36 #30
title_pad = 24 #20

# Axes Parameters
axes = plt.gca()
axes.tick_params(axis='both') 

# Axes ---> Linewidth
for axis in ['top','bottom','left','right']:
    axes.spines[axis].set_linewidth(2.2)
    
# Axes ---> Ticks
axes.xaxis.set_minor_locator(AutoMinorLocator(5)) 
plt.tick_params(which='minor', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=6)
axes.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.tick_params(which='minor', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=6)
plt.tick_params(which='major', direction='in', labelsize=label_size, top=True, right=True, width=1.2, length=12)

# Grid
#plt.grid(True)
    
#############################################
## Plot specific settings and plot data
#############################################

# Data plot

hist_rng_min = 150 #min(Delta_t)
hist_rng_max = 500 #max(Delta_t)
rng_scale = 1
bins = np.linspace(hist_rng_min, hist_rng_max, int((hist_rng_max - hist_rng_min)/10/rng_scale)) # Delta_t only has elements which are multiples of 10

time_diff_hist, binEdges = np.histogram(Delta_t, bins=bins)
bincenters = 0.5 * (binEdges[1:] + binEdges[:-1])

# Fit exponential

xdata=bincenters[1:-1]
ydata=time_diff_hist[1:-1]

popt, pcov = curve_fit(expo, xdata, ydata, p0=[2000, 0.01])
kr_halflife = ((1/popt[1])*np.log(2))
perr = np.sqrt(np.diag(pcov))
error = (np.sqrt((-1/(popt[1]**2)*(np.sqrt(pcov[1,1])))**2))

chi_squared =  (np.sum((ydata - expo(xdata, *popt))**2 / (np.sqrt(ydata))**2))
Reduced_chi_squared = ( chi_squared / len(xdata))


plt.errorbar(xdata, ydata, yerr=np.sqrt(ydata), 
             xerr=(xdata[-1]-xdata[0])/len(xdata)/np.sqrt(12), fmt = 'o', 
             label = 'Measurement Data')

plt.plot(xdata, expo(xdata, *popt), linewidth = 2.2, zorder = 0,
         label = 'Exponetial Fit')


plt.xlabel('Time Difference S1 - S2 [ns]', fontsize = label_size, labelpad = label_pad)
plt.ylabel('Counts', fontsize = label_size, labelpad = label_pad)
#plt.title('Half-Life', fontsize = title_size, pad = title_pad)

# Legend
leg = plt.legend(loc=1, bbox_to_anchor=(0.99,0.99), fontsize=0.9*label_size, frameon=True, shadow=False, edgecolor='black', fancybox=False)
for line in leg.get_lines(): # enumerated
    line.set_linewidth(4)
leg.get_frame().set_linewidth(2.5)
leg._legend_box.align = "left"

fig.savefig(filenameprefix+"_Kr_lifetime.png", bbox_inches='tight')

plt.show()
```
